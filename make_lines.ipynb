```python
import os
import re
```

```python
data_dir = "data/txt/"
output_dir = "data/lines/"
```

```python
data_list = os.listdir(data_dir)
```

```python
data_list
```

```python
for name in data_list:
    if name.endswith(".txt"):
        path = os.path.join(data_dir, name)
        with open(path, "rb") as f:
            lines = f.readlines()
        for line in lines:
            line = line.decode("utf-8")
            char_name = line.split("「")[0]
            char_line = re.split(r"「|」", line)[1]
            char_line = re.sub(r"\(.+?\)", "", char_line)
            print(char_name, char_line)
            if os.path.exists(os.path.join(output_dir, char_name+".txt")):
                with open(os.path.join(output_dir, char_name+".txt"), "a") as f:
                    f.write(char_line+"\n")
            else:
                with open(os.path.join(output_dir, char_name+".txt"), "w") as f:
                    f.write(char_line+"\n")
```
